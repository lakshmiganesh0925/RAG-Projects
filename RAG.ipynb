```python
from google.colab import userdata
api_key=userdata.get('GEMINI_API_KEY')
```

```python
from langchain_community.document_loaders import PyPDFLoader
loader =PyPDFLoader('/content/2210.02928v2.pdf')
doc=loader.load()
```

```python
doc[0].metadata
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200,
)
```

```python
all_splits=text_splitter.split_documents(doc)
```

```python
all_splits
```

```python
!pip install langchain-huggingface sentence_transformers
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
embedding_model = HuggingFaceEmbeddings(
    model='sentence-transformers/all-mpnet-base-v2'
)
```

```python
!pip install langchain-chroma
```

```python
from langchain_core.vectorstores import InMemoryVectorStore

vectorstore=InMemoryVectorStore(
    embedding_model
)
```

```python
vectorstore.add_documents(documents=all_splits)
```

```python
!pip install langchain_google_genai
```

```python

from langchain_google_genai import GoogleGenerativeAI
model= GoogleGenerativeAI(model="models/text-bison-001", google_api_key=api_key)
```

```python
def retrieve_context(query: str, k: int = 2):
  retrieved_docs = vectorstore.similarity_search(query, k=k)
```

```python
def docu_chat(user_query):
  context, source_docs = retrieve_context(user_query, k=2)
  system_message = f"""You are a helpful chatbot.
                     Use only the following pieces of context to answer the
                     question. Don't makeup any new information: {context} """

  messages = [
    {"role": "system", "content": system_message},
    {"role": "user", "content": user_query}
  ]
  response = model.invoke(messages)
  return {
  "answer": response.content,
  "source_documents": source_docs,
  "context_used": context
}
result = docu_chat( "Explain what is the use of decoders in transformers?")
print(result)
print(result["answer"])
```
