```python
from google.colab import userdata
api_key=userdata.get('GEMINI_API_KEY')
```

```python
import os
os.environ['GEMINI_API_KEY']=api_key
```

```python
from langchain_community.document_loaders import WebBaseLoader
from langchain_community.vectorstores import FAISS
```

```python
loader=WebBaseLoader('https://en.wikipedia.org/wiki/State_of_the_Union#:~:text=Though%20the%20language%20of%20the,as%20late%20as%20March%207',encoding='utf8')
```

```python
document=loader.load()
```

```python
print(document[0].page_content)
```

```python
!pip install langchain-text-splitters faiss-cpu
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
text_splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
```

```python
text_chunks=text_splitter.split_documents(document)
```

```python
text_chunks
```

```python
from langchain_google_genai.embeddings import GoogleGenerativeAIEmbeddings
from langchain_community.vectorstores import FAISS
```

```python
! pip install -qU langchain langchain-huggingface sentence_transformers
```

```python
from langchain_huggingface import HuggingFaceEmbeddings

embedding_model = HuggingFaceEmbeddings(
  model_name = "sentence-transformers/all-mpnet-base-v2"
)
```

```python
vectorstore=FAISS.from_documents(text_chunks,embedding=embedding_model)
```

```python
retriever = vectorstore.as_retriever()
```

```python
from langchain_core.prompts import ChatPromptTemplate
```

```python

template="""You are an assistant for question-answering tasks.
Use the following pieces of retrieved context to answer the question.
If you don't know the answer, just say that you don't know.
Use ten sentences maximum and keep the answer concise.
Question: {question}
Context: {context}
Answer:
"""
```

```python
prompt=ChatPromptTemplate.from_template(template=template)
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser
```

```python
output_parser =StrOutputParser()
```

```python
llm_model= ChatGoogleGenerativeAI(model='gemini-2.5-flash',api_key=api_key)
```

```python
rag_chain=(
    {"context":retriever,"question":RunnablePassthrough()}
    |prompt
    |llm_model
    |output_parser
)
```

```python
rag_chain.invoke("What action is the U.S. taking to address rising gas prices?")
```
